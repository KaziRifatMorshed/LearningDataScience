```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate Fibonacci numbers
def fibonacci(n):
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib

# Generate first 100 Fibonacci numbers
fib_numbers = fibonacci(100)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(range(100), fib_numbers, 'b-', label='Fibonacci Sequence')
plt.title('First 100 Fibonacci Numbers')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

# Use scientific notation for y-axis due to large numbers
plt.yscale('log')

plt.show()

# Print the last number to show the magnitude
print(f"The 100th Fibonacci number is: {fib_numbers[-1]}")
```
